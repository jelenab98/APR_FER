```python
def correct_hits(d, epsilon=1e-6):
    i = 0
    for j in d:
        if j <=epsilon:
            i += 1
    return i
```

# Drugi zadatak
***
Ispitivanje povećanja dimenzionalnosti i uspješnosti GA za F6 i F7

Dimenzije su (1, 3, 6, 10), bilježe se vrijednosti funkcije kazne u najboljim jedinkama što se kasnije prikazuje na grafu.

```python
dimensions = (1, 3, 6, 10)
n_iters = 5
functions = (6, 7)
data = {6:[], 7:[]}
for f in functions:
    for d in dimensions:
        tmp_val = []
        for i in range(n_iters):
            if f == 6:
                fn = Functions.F6()
                v_pop = 1000
                p = 0.3
            else:
                fn = Functions.F7()
                v_pop = 500
                v = 0.5
            ga = GeneticAlgorithm.DiscreteGA(function=fn, crossovers=["whole_arithmetic", "discrete", "single", 
                                             "simple", "one"],mutations=["uniform", "gauss"], max_iter=1e5, 
                                             gene_size=d, interval=(-50, 150), population_size=v_pop, p_mutation=p, 
                                             epsilon=1e-6, printing=False)
            print("\nFunction: {}, Dimension: {}".format(f, d))
            ga.calculate()
            tmp_val.append(ga.min_error)
        data[f].append(tmp_val)
```

```python
with open("zad2.json", 'w') as f:
    json.dump(data, f)

x_labels = []
plt.figure(figsize=(15, 10))
plt.subplot(2, 1, 1)
dims = [[1, 1, 1, 1, 1], [3, 3, 3, 3, 3], [6, 6, 6, 6, 6], [10, 10, 10, 10, 10]]
for i in range(len(dims)):
    plt.scatter(dims[i], data[6][i])
    x_labels.append("d={}, pogođeno={}".format(dimensions[i], correct_hits(data[6][i])))
plt.xticks(dimensions, x_labels);
plt.title("Funkcija 6 uz vel_pop=1000, p_mut=0.3")

x_labels = []
plt.subplot(2, 1, 2)
for i in range(len(dims)):
    plt.scatter(dims[i], data[7][i])
    x_labels.append("d={}, pogođeno={}".format(dimensions[i], correct_hits(data[7][i])))
plt.xticks(dimensions, x_labels);
plt.title("Funkcija 7 uz vel_pop=500, p_mut=0.5");
```

# Treći zadatak
***
Usporedba prikaza rješenja za funkcije 6 i 7. Usporedba se provodi po prethodno navedenim parametrima uz korištenje preciznosti od 1e-4 te veličine jedinki 3 i 6.

```python
functions = [7]
dimensions = (3, 6)
n_iters = 10
data_fp[7] = []
data_bin[7] = []
#data_fp = {6:[], 7:[]}
#data_bin = {6:[], 7:[]}
for f in functions:
    if f == 6:
        fn = Functions.F6()
        v_pop = 1000
        p_mut = 0.3
    else:
        fn = Functions.F7()
        v_pop = 500
        p_mut = 0.5
    for d in dimensions:
        tmp_data_fp = []
        tmp_data_bin = []
        for i in range(n_iters):
            ga_fp = GeneticAlgorithm.DiscreteGA(function=fn, crossovers=["whole_arithmetic", "discrete", "single", 
                                 "simple", "heuristic"],mutations=["uniform", "gauss"], max_iter=1e5, gene_size=d, 
                                 interval=(-50, 150), population_size=v_pop, p_mutation=p_mut, 
                                 epsilon=1e-6, printing=False)
            
            ga_bin = GeneticAlgorithm.BinaryGA(function=fn, crossovers=["single", "multi", "uniform"], 
                                           mutations=["uniform", "simple"], max_iter=1e5, gene_size=d, 
                                           interval=(-50, 150), population_size=v_pop, p_mutation=p_mut, 
                                           epsilon=1e-6, precision=4, printing=False)
            print("\nFunction: {}, Dimension: {}".format(f, d))
            ga_fp.calculate()
            ga_bin.calculate()
            tmp_data_bin.append(ga_bin.min_error)
            tmp_data_fp.append(ga_fp.min_error)
        data_fp[f].append(tmp_data_fp)
        data_bin[f].append(tmp_data_bin)
```

```python
with open("zad3_fp.json", 'w') as f:
    json.dump(data_fp, f)
    
with open("zad3_bin.json", 'w') as f:
    json.dump(data_bin, f)
```

Idealni parametri za funkciju 6.

- vjerojatnost mutacije = 0.01, 0.3, 0.5, 0.9

```python
vel_pop = (30, 50, 100, 300, 1000)
p_mut = (0.01, 0.3, 0.5, 0.9)
n_iters = 10
gene_size = 2
fn = Functions.F6()
datas = {30:{0.01:[], 0.3:[], 0.5:[], 0.9:[]}, 
         50:{0.01:[], 0.3:[], 0.5:[], 0.9:[]}, 
         100:{0.01:[], 0.3:[], 0.5:[], 0.9:[]}, 
         300:{0.01:[], 0.3:[], 0.5:[], 0.9:[]}, 
         1000:{0.01:[], 0.3:[], 0.5:[], 0.9:[]}}

for v_pop in vel_pop:
    for p in p_mut:
        print("\nV_pop: {}, p:{}".format(v_pop, p))
        for i in range(n_iters):
            ga_fp = GeneticAlgorithm.DiscreteGA(function=fn, crossovers=["whole_arithmetic", "discrete", "single", 
                     "simple", "one"],mutations=["uniform", "gauss"], max_iter=1e5, gene_size=gene_size, 
                     interval=(-50, 150), population_size=v_pop, p_mutation=p, 
                     epsilon=1e-6, printing=False)
            ga_fp.calculate()
            datas[v_pop][p].append(ga_fp.min_error)
```

# Peti zadatak
***
Usporedba uspješnosti GA u ovisnosti o veličini parametra k kod k-turnirske implementacije. 


```python
data_k = {3:[], 5:[], 7:[], 9:[], 11:[], 13:[]}
k = (3, 5, 7, 9, 11, 13)
n_iters = 10
fn = Functions.F6()
for k_i in k:
    print("\nk = {}".format(k_i))
    for i in range(n_iters):
        ga_3 = GeneticAlgorithm.DiscreteGA(function=fn, crossovers=["whole_arithmetic", "discrete", "single", 
                             "simple", "heuristic"],mutations=["uniform", "gauss"], max_iter=1e5, gene_size=2, 
                             interval=(-50, 150), population_size=1000, p_mutation=0.3, k=k_i, 
                             epsilon=1e-6, printing=False)
        ga_3.calculate()
        data_k[k_i].append(ga_3.min_error)
```

```python
with open("zad5_2.json", 'w') as f:
    json.dump(data_k, f)
```

```python
plt.figure(figsize=(20, 8))
plt.boxplot(data_k.values())
labels = ["k={}, hits={}".format(i, correct_hits(data_k[i])) for i in k]
plt.xticks(range(1, len(data_k) + 1), labels)
plt.title("Usporedba uspješnosti GA za različiti k");
```
